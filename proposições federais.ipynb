```python
#df = pd.DataFrame(documents)
#df.columns = ["Titulo", "Conteudo"]
#df
df = pd.read_csv("proposicoes.csv", sep = ";")
df_tratado = df.drop(columns=["uriProposicao","numero", "ano", "codTema","relevancia"])

# splitting dataframe in a particular size
df_split = df_tratado.sample(frac=0.05,random_state=200)
df_split.reset_index()
df_split
```

```python
model = "models/embedding-001"
```

```python
def embed_fn(siglaTipo, tema):
  return genai.embed_content(model=model,
                                 content=tema,
                                 title=siglaTipo,
                                 task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df_split["Embeddings"] = df_split.apply(lambda row: embed_fn(row["siglaTipo"], row["tema"]), axis=1)
df_split
```

```python
df_split
```

```python
def gerar_e_buscar_consulta(consulta, base, model):
  embedding_da_consulta = genai.embed_content(model=model,
                                 content=consulta,
                                 task_type="RETRIEVAL_QUERY")["embedding"]

  produtos_escalares = np.dot(np.stack(df_split["Embeddings"]), embedding_da_consulta)

  indice = np.argmax(produtos_escalares)
  return df_split.iloc[indice]["tema"]
```
